```python
import pandas as pd
```

```python
df1 = pd.read_csv("siteless_rarefaction_data_202109_genera.csv")
df2 = pd.read_csv("202109_site_list.csv")
```

```python
df1.shape
```

```python
df2.shape
```

```python
list1 = []
for items in df2.columns:
    list1.append(items)
```

```python
print(list1)
```

```python
df2.loc[-1] = list1
df2.index = df2.index + 1  # shifting index
df2 = df2.sort_index()  # sorting by index
df2.head()
```

```python
df2.columns[1]
```

```python
df2.rename(columns = {df2.columns[0]:'meta_id', df2.columns[1]: 'sample_id', df2.columns[2]: 'site'}, inplace = True)
```

```python
df2.head()
```

```python
df3 = (pd.merge(df1, df2, left_on='samples', right_on='meta_id', how='left').drop('meta_id', axis=1))
```

```python
df3.head()
```

```python
cols = list(df3.columns.values)
print(cols)
```

```python
df3.head()
```

```python
df3 = df3.drop('Unnamed: 0', axis = 1)
```

```python
df3.head()
```

```python
df3.insert(1, 'sample_id', df3.pop('sample_id'))
df3.insert(2, 'site', df3.pop('site'))
df3.head()
```

```python
df3.to_csv('full_202109_genera_with_site.csv', index=False)
```
