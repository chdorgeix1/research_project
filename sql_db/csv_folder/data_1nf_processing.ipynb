```python
ldf2['sample_id'] = ldf2['Plot'].str.cat(ldf2['Plant'], sep = '')
#df['Combined_Column'] = df['Column1'].str.cat(df['Column2'], sep=' ')

ldf2.head()
```

```python
merged_df.columns
```

```python
merged_df = merged_df.drop(['site_id',  'date', 'Plot',
       'Plant'], axis = 1)
```

```python
merged_df.head()
```

```python
merged_df = merged_df.drop_duplicates(subset=['sample_id'])
merged_df.shape
```

```python
merged_df = merged_df.rename(columns={'Ht(cm)': 'plant_ht', '# Leaves': 'num_leaves', 'Length LL (cm)': 'longest_leaf_cm',
                                     'Width LL (cm)': 'widest_leaf_cm', 'Stem Diameter (mm)': 'stem_diam_mm',
                                     '# Inflorescences': 'num_inflores', '# Flowers Per (x2)': 'num_flowers',
                                     'Herbivory': 'herbivory', 'Species': 'species'})
```

```python
merged_df.columns
```

```python
merged_df['Species_x'].dropna().shape
```

```python
merged_df = merged_df.rename(columns={'Species_x': 'observed_species','Species_y': 'casey_species', 
                                      'length_leaf': 'casey_length', 'width_leaf': 'casey_width'})
```

```python
merged_df.columns
```

```python
merged_df.to_csv('combined_morphology_csv.csv')
```

```python
sdf = pd.read_csv('./csv_folder/soil_nutrients.csv')
sdf.head()
```

```python
soil_list = sdf['sample_id'].tolist()
temp_list = []
for i in soil_list:
    if i[0] == '8':
        temp_list.append(i)
for i in temp_list:
    sdf.drop(sdf[(sdf['sample_id'] == i)].index, inplace=True)
sdf.drop(sdf[(sdf['sample_id'] == 'BLDRV1-')].index, inplace=True)
sdf.drop_duplicates(subset="sample_id",
                         keep='first', inplace=True)
soil_list = sdf['sample_id'].tolist()
for i in range(len(soil_list)):
    if soil_list[i][0] == '1':
        soil_list[i] = soil_list[i][1:]
    if soil_list[i][0] == ' ':
        soil_list[i] = soil_list[i][1:]
print(len(soil_list))
```

```python
sdf['sample_id'] = soil_list
```

```python
sdf = sdf.drop('site', axis = 1)
sdf.head()
```

```python
newdf = sdf.melt('sample_id', var_name='nutrient', value_name='nutrient_value')
```

```python
newdf.to_csv("final_soil.csv")
```

```python
pmdf = pd.read_csv('plant_microbe_genus.csv')
pmdf.head()
```
